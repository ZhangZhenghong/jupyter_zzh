```python
from time import time
import functools

#装饰器写法
def outer(func):
    @functools.wraps(func)
    def wrapper(*args, **kwargs):
        print('start', func.__name__)
        start = time()
        func(*args, **kwargs)
        end = time()
        print('total', str(end-start) )
    return wrapper
        
class Node():
    def __init__(self,item):
        self.item = item
        self.lchild = None
        self.rchild = None

class Mytree():
    def __init__(self):
        self.root = None

    def add(self, item):
        node = Node(item)
        if self.root is None:
            self.root = node
            return
        else:
            queue = [self.root]
            while queue:
                cur_node = queue.pop(0)
                if cur_node.lchild is None:
                    cur_node.lchild = node
                    return
                else:
                    queue.append(cur_node.lchild)
                if cur_node.rchild is None:
                    cur_node.rchild = node
                    return
                else:
                    queue.append(cur_node.rchild)
                    
    @outer
    def travel(self): 
        '''
        采用队列来遍历树，先遍历左子树，如果不为空，加到队列，再遍历右子树，如果不为空，加到队列，
        '''
        queue = [self.root]
        while queue:
            cur_node = queue.pop(0)
            print(cur_node.item)
            if cur_node.lchild is not None:
                queue.append(cur_node.lchild)
            if cur_node.rchild is not None:
                queue.append(cur_node.rchild)
        return

    def xian(self,root):
        '''
        递归调用自身
        '''
        root = root
        if root is None:
            return
        else:
            print(root.item, end='')
            self.xian(root.lchild)
            self.xian(root.rchild)

    def hou(self, root):
        root = root
        if root is None:
            return
        else:
            self.hou(root.lchild)
            self.hou(root.rchild)
            print(root.item, end='')
            


if __name__ == '__main__':
    tree = Mytree()
    tree.add(0)
    tree.add(1)
    tree.add(2)
    tree.add(3)
    tree.add(4)
    tree.add(5)
    tree.add(6)
    tree.add(7)

    tree.travel()
    tree.xian(tree.root)
    print("\n")
    tree.hou(tree.root)
```
